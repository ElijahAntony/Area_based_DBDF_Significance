```python
import pingouin as pg
import pandas as pd
import numpy as np
```

```python
df_ab = pd.read_csv('./airflow_significance/anova_all_areas_final.csv')
```

```python
df_ab.head()
```

```python
areas=df_ab['area_name'].drop_duplicates()
areas=areas.dropna()
areas
```

```python
an=pd.DataFrame()
for ar in areas:
    a=df_ab[df_ab['area_name']==ar]
    post_hoc_results = pg.welch_anova(a[['variant', 'orders']], dv='orders', between='variant').copy()
    post_hoc_results['area_name']=ar
    an=pd.concat([an,post_hoc_results])
```

```python
an
```

```python
gh=pd.DataFrame()
for ar in areas:
    b=df_ab[df_ab['area_name']==ar]
    pgh = pg.pairwise_gameshowell(b[['variant', 'orders']], dv='orders', between='variant',
                                                               effsize='hedges').copy()
    pgh['area_name']=ar
    gh=pd.concat([gh,pgh[0:5]])
```

```python
gh
```

```python

gh['p_corr'] = pg.multicomp(gh['pval'].to_numpy(), alpha=0.05, method='bonf')[1]
gh.head()
```

```python
from pathlib import Path  
filepath = Path('./airflow_significance/anova_op2.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
an.to_csv(filepath)  
```
